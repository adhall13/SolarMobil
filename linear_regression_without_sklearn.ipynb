```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('Fish[1].csv') 

data = df[['Length1', 'Weight']].dropna()
x = data['Length1']
y = data['Weight']

x_mean = np.mean(x)
y_mean = np.mean(y)

# Calculate slope (m) and intercept (b)
m = np.sum((x - x_mean) * (y - y_mean)) / np.sum((x - x_mean)**2)
b = y_mean - m * x_mean

# Predictions
y_pred = m * x + b

# R^2
rss = np.sum((y - y_pred)**2)
tss = np.sum((y - y_mean)**2)
r2 = 1 - (rss / tss)

# Result
print(f"Slope (m): {m:.4f}")
print(f"Intercept (b): {b:.4f}")
print(f"R2 Score: {r2:.4f}")

plt.scatter(x, y, alpha=0.5, label='Data')
plt.plot(x, y_pred, color='red', label='Regression Line')
plt.xlabel('Length')
plt.ylabel('Weight')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('Fish[1].csv')

# removing any rows with missing values to ensure the math works
df_clean = df[['Length1', 'Weight']].dropna()
x = df_clean['Length1']
y = df_clean['Weight']

# R^2
x_mean, y_mean = np.mean(x), np.mean(y)
m = np.sum((x - x_mean) * (y - y_mean)) / np.sum((x - x_mean)**2)
b = y_mean - m * x_mean

y_pred = m * x + b
rss = np.sum((y - y_pred)**2)
tss = np.sum((y - y_mean)**2)
r2 = 1 - (rss / tss)

# graph 
plt.figure(figsize=(10, 6))
sns.set_theme(style="whitegrid")

# sns.regplot draws the scatter points and the regression line automatically
sns.regplot(
    data=df_clean, 
    x='Length1', 
    y='Weight', 
    scatter_kws={'alpha': 0.6, 'color': 'steelblue'},
    line_kws={'color': 'darkred', 'label': f'Regression Line (R²={r2:.3f})'}
)

# format
plt.title('Linear Regression: Fish Length vs Weight', fontsize=16)
plt.xlabel('Length 1 (cm)', fontsize=12)
plt.ylabel('Weight (g)', fontsize=12)
plt.legend()

# Print statistical results to the console
print(f"--- Model Results ---")
print(f"Slope (m):     {m:.4f}")
print(f"Intercept (b): {b:.4f}")
print(f"R2 Score:      {r2:.4f}")

plt.show()
```
